<b>Задание. <br>Урок 9. Детальная подготовка инициатив по оптимизации.</b><br>
Вы работаете аналитиком в HR департаменте маркетингового агентства Х.<br> 
В агентстве за последний год снизились бизнес результаты, и руководство предполагает, что это <br>
произошло, как результат снижения удовлетворенности сотрудников.<br> 
Вы провели опрос сотрудников, проанализировали результаты и увидели, что сотрудники отмечают<br> 
изменение к худшему корпоративной культуры, переработки, высокий уровень стресса в командах,<br>
что уже привело к увольнению части сотрудников, выгоранию ряда сотрудников,<br> 
снижению вовлеченности и мотивации сотрудников, снижение уровня креативности сотрудников,<br>
а также к ухудшению взаимоотношений в командах.<br> 
    
Предложите 3 инициативы по улучшению ситуации и приоритизируйте их по RICE,<br> 
опишите методологию оценки по RICE<br> 
(что значит ваша шкала оценки, оцениваете ли вы от 1 до 5 баллов, от 0 до 100% или еще как-то).

В результате проведенных опросов были выявлены последствия, а не причины.<br>
Стало только понятней, что причины снижения бизнес-результатов могут, но *не обязаны*, <br>
лежать в организации труда и выявили плохое состояние рабочей атмосферы среди сотрудников.<br>
Однако, абсолютно не ясно, какие измения в работе сотрудников привели к текущей ситуации.<br>
Пока, бесспорно только одно: **сотрудники работают без прежнего энтузиазма, работа стала втягость,<br> собачатся друг с другом и эмоционально увольняются, "хлопнув дверью"**.<br>
Типо **_"Да пошли вы все куда подальше"_**.<br>

Ситуация близка к катастрофе и развалу компании.<br>
**Требуются срочные меры!**<br>
Но неизвестно какие конкретно.

Поэтому, я бы, прежде всего, провел личные встречи с сотрудниками,<br>
на которых максимально детализировал бы вопросы по возможным причинам проблем,<br>
таким, как:
+ поставленные задачи слишком просты и не соответсвуют квалификации сотрудника; 
+ сроки выполнения работы слишком маленькие;
+ сложность задачи превосходит квалификацию сотрудника;
+ неправильная организация рабочего времени(нет времени на отдых);
+ слишком большое количество задач для одного сотрудника;
+ тема задачи не вызывает интерес у сотрудника;
+ неправильная организация рабочего места (нет возможности сосредоточиться);
+ заработная плата не соответствует уровню задач;
+ нет системы премий или просто поощрений;
+ сотрудники плохо знают друг друга и сложно коммуницируют;
+ сотрудники не понимают главную задачу компании и важность их вклада в общий результат;
+ техническое состояние предоставленного компанией оборудования слишком старое<br>
и не позволчет качественно выполнить работу;
+ и, возможно, ещё что…

Но, так как в задании уже сейчас хотят приоритезацию по RICE, что ж.<br>
Но инициатив у меня будет шесть.<br>
