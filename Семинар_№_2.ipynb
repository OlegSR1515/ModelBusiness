**Основными**, или **операционными** бизнес-процессами, называют процессы,<br>
от которых напрямую зависит прибыль компании.<br>
Например :

 + Продажа;
 + Маркетинг;
 + Производство.

**Вспомогательные** процессы не добавляют ценности продукции и являются затратными по своей сути.<br>
К таким процессам обычно относятся: 

- Подготовка кадров; 
- Сервисное обслуживание оборудования; 
- Обеспечение связью;
- IT – обеспечение; 
- Административно – хозяйственное деятельность.
